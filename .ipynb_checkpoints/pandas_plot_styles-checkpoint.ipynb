```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config inlineBackend.figure_format = 'retina'   #2x
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/prasertcbs/tutorial/master/mpg.csv')
df.sample(10)
```

```python
df.drv.value_counts()
```

```python
df.drv.value_counts().plot.barh();
```

## GGPlot

```python
plt.style.use('ggplot');
df.drv.value_counts().plot.barh();
```
