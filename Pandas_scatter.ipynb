```python
df.plot.scatter(x='displ',y='cty');
```

```python
df.plot.scatter(x='displ',y='cty',
                grid = True,
                c = df.drv.map({'4':'green','f':'blue','r':'red'}),
                s = df.displ*5,
                title='City miles per gallon vs Displacement');
```

```python
df.plot.scatter(x='displ',y='cty',
                grid = True,
                c = df.drv.map({'4':'green','f':'blue','r':'red'}),
                s = df.cyl*5,
                alpha = .6,
                title='City miles per gallon vs Displacement');
```
